```python
# Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine.
'''
 The key features of the wine quality data set are('fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol').



1. **Fixed Acidity**: This feature represents the concentration of non-volatile acids in the wine.
Acidity is a crucial factor in determining the overall taste and balance of a wine. Wines with higher fixed acidity tend to be more vibrant and refreshing,
contributing to a better quality perception.

2. **Volatile Acidity**: Volatile acidity is the measure of volatile acids (such as acetic acid) in the wine.
 Excessive volatile acidity can lead to off-putting vinegar-like aromas and flavors, indicating poor quality.

3. **Citric Acid**: Citric acid is a natural component found in grapes and can contribute to the overall freshness and fruitiness of the wine.
 Proper balance of citric acid can enhance the quality of wine by adding complexity and bright flavors.

4. **Residual Sugar**: This refers to the amount of sugar left in the wine after fermentation.
The level of residual sugar can impact the perceived sweetness of the wine. Finding the right balance is crucial for achieving the desired taste profile,
and it can influence the perceived quality.

5. **Chlorides**: Chlorides, often in the form of salt, can affect the taste and mouthfeel of wine. Higher chloride concentrations can impart a salty taste, which may not be desirable and could negatively impact the quality.

6. **Free Sulfur Dioxide**: Sulfur dioxide is commonly used as a preservative in winemaking.
The amount of free sulfur dioxide can impact the wine's stability and longevity. Too much or too little can affect the wine's quality and shelf life.

7. **Total Sulfur Dioxide**: This includes both free and bound forms of sulfur dioxide.
 It's another measure of the wine's preservation and stability. A balance is necessary to maintain wine quality without negatively affecting the taste.

8. **Density**: Density is a measure of the wine's mass per unit volume.
It can provide insights into the concentration of dissolved solids in the wine, which can influence mouthfeel and texture.
High density might indicate higher quality due to increased flavor concentration.

9. **pH**: pH indicates the acidity or basicity of the wine. It's a crucial factor in influencing the wine's overall balance and stability.
Wines with appropriate pH levels tend to be more harmonious and of higher quality.

10. **Sulphates (Sulfites)**: Sulphates are compounds that can be added to wine as preservatives. They can prevent spoilage and oxidation.
Proper use of sulphates can contribute to maintaining the wine's quality over time.

11. **Alcohol**: Alcohol content significantly affects the wine's body, texture, and perceived warmth.
Finding the right alcohol level for a particular wine style is important for achieving a balanced and enjoyable product.

12. **Quality**: This is the target variable, representing the perceived quality of the wine.
It's often a subjective measure determined by experts or consumers. The other features in the dataset are used to predict this quality score.

In summary, each feature in the wine quality dataset contributes to different aspects of the wine's sensory experience and chemical composition.
Analyzing these features collectively can help predict and understand the overall quality of the wine,
ensuring it meets the desired flavor profile and appeals to consumers' preferences.'''
```

```python
# Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features.
# What is the minimum number of principal components required to explain 90% of the variance in the data?

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the wine quality dataset
df = pd.read_csv('winequality-red.csv')

# Separate features and target
X = df.drop(columns=['quality'])
y = df['quality']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA()
pca.fit(X_scaled)

# Calculate explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Calculate cumulative explained variance
cumulative_explained_variance = explained_variance_ratio.cumsum()

# Determine the number of components for 90% variance
n_components_90 = np.argmax(cumulative_explained_variance >= 0.9) + 1

print(f"Number of components required for 90% variance: {n_components_90}")
```
